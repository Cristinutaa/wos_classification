```python
import pandas as pd

# Define lists to store parsed data
discipline1_list = []
discipline2_list = []
score_list = []
discipline1_for_discipline1_list = []
discipline1_score = 0
# Open and read the file
with open('results_nn_100000.txt', 'r') as file:
    discipline1 = ''
    for line in file.readlines():
        if line.startswith('Discipline'):
            # Extract discipline1 and its score for this block
            discipline1 = line.split(':')[1].strip()
            if discipline1_score: 
                discipline1_for_discipline1_list.extend([discipline1_score for i in range(42)])
        elif line.startswith('  '):
            # Extract discipline2 and score
            parts = line.strip().split(': ')
            discipline2 = parts[0].strip()
            score = float(parts[1].strip())
            
            # Append the data to lists
            discipline1_list.append(discipline1)
            discipline2_list.append(discipline2)
            score_list.append(score)
            if discipline1 == discipline2: 
                discipline1_score = score
    discipline1_for_discipline1_list.extend([discipline1_score for i in range(42)])

# Create a DataFrame
df = pd.DataFrame({
    'Discipline1': discipline1_list,
    'Discipline2': discipline2_list,
    'Score': score_list,
    'Discipline1Value': discipline1_for_discipline1_list
})

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import pandas as pd

# Load the DataFrame
df = pd.read_csv('data.csv')  # Make sure to adjust the path to your CSV file

# Initialize an empty list to store the final data
final_data = []

# Iterate over each unique discipline in Discipline1
for discipline in df['Discipline1'].unique():
    discipline_data = df[df['Discipline1'] == discipline]
    S = discipline_data['Discipline1Value'].iloc[0]

    # Calculations
    disciplines_greater_than_S_minus_1_list = discipline_data[discipline_data['Score'] > S - 1]
    disciplines_greater_than_S_minus_1 = len(disciplines_greater_than_S_minus_1_list)
    disciplines_score_above_1 = len(discipline_data[discipline_data['Score'] > 1])

    # Exclude the current discipline from the first 3 calculations
    if discipline in disciplines_greater_than_S_minus_1_list['Discipline2'].values:
        disciplines_greater_than_S_minus_1 -= 1
        disciplines_greater_than_S_minus_1_list = disciplines_greater_than_S_minus_1_list[disciplines_greater_than_S_minus_1_list['Discipline2'] != discipline]['Discipline2'].tolist()
    else:
        disciplines_greater_than_S_minus_1_list = disciplines_greater_than_S_minus_1_list['Discipline2'].tolist()

    rank_score_more_than_S_minus_1 = len(df[(df['Discipline2'] == discipline) & (df['Score'] > df['Discipline1Value'] - 1)])
    rank_score_more_than_S_minus_1_list = df[(df['Discipline2'] == discipline) & (df['Score'] > df['Discipline1Value'] - 1)].Discipline1.values
    rank_score_above_1 = len(df[(df['Discipline2'] == discipline) & (df['Score'] > 1)])

    final_data.append({
        'Discipline': discipline,
        'Disciplines > (S-1)': disciplines_greater_than_S_minus_1,
        'Disciplines with Score > 1': disciplines_score_above_1,
        'Nb of disciplines where this discipline is in the rank with score more than (S-1)': rank_score_more_than_S_minus_1,
        'Nb of disciplines where this discipline is in the ranking with score above 1': rank_score_above_1,
        'Disciplines List > (S-1)': disciplines_greater_than_S_minus_1_list,
        'Disciplines List score Sxa > Sx': rank_score_more_than_S_minus_1_list
    })

# Create a DataFrame from the final data list
final_df = pd.DataFrame(final_data)

# Display the DataFrame
final_df
```

```python
final_df.to_csv('info_table.csv', index=False)
```
